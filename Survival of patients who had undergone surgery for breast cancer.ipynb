# CDF AND PDF
- IN ALL THE BELOW GRAPH ORANGE LINE REPRESENT CDF AND BLUE LINE REPRESENT CDF

```python
counts, bin_edges=np.histogram(cancer['aop'],bins=10,density=True)
pdf=counts/(sum(counts))
#pdf
print(pdf);
print(bin_edges);
#cdf
cdf=np.cumsum(pdf)
plt.ylabel('percentages')
plt.xlabel('age of operation')
plt.title('PDF AND CDF OF AGE OF OPERATION OF PATIENT')
plt.plot(bin_edges[1:],pdf);#pdf graph
plt.plot(bin_edges[1:],cdf);#cdf graph
```

- IN THE ABOVE GRAPH ORANGE LINE IS THE CDF AND BLUE LINE IS THE PDF
- ORANGE LINE IS THE SUM OF EVERY NEW VALUE WHICH WE CAN CALL AS A CUMMILATIVE SUM "NP.CUMSUM"
- WE CAN READ IT AS "THERE IS 43% OF OPERATIONS IN WHICH PATIENT AG IS LESS THAN OR EQUAL TO 50
- POINT TO REMEMBER: CDF IS ALWAYS HIGHER THAN THE PDF

```python
counts, bin_edges=np.histogram(cancer['yop'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf=np.cumsum(pdf)
plt.ylabel('percentage')
plt.xlabel('year of operation')
plt.title('pdf and cdf of year of operation of patients')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf);
```

```python
counts,bin_edges=np.histogram(cancer['pand'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf=np.cumsum(pdf)
plt.ylabel('percentage')
plt.xlabel('positive auxillary nodes')
plt.title('pdf and cdf of the positive auxillary nodes detected')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf);
```

```python

counts,bin_edges=np.histogram(cancer['ss'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf=np.cumsum(pdf)
plt.ylabel('percentage')
plt.xlabel('survival status')
plt.title('pdf and cdf of survival status of the patients')
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf);
```

```python
sns.boxplot(x='ss',y='yop',data=cancer)
plt.title('boxplot between survival status and year of operation')
plt.plot();
```

It is the another way of visualizing the 1-d scatter plot
Boxplot mainly helps in measuring the value minimum and maximum of any element present in the dataset
from the above plot we can say that patients who survived more than 5 years after operation have min age of 58 during year of operation and maximum is above 68. similar in condition the patients who didn't survive more than 5 years after operation and these minimum and maximum value indicators are called whiskers

```python
sns.violinplot(x='ss',y='yop',data=cancer)
plt.title('violinplot between year of operation and survival status')
plt.plot();
```

- Violinplot is the combination of histogram and boxplot.
- The black portion inside the random shape is a boxplot and the white dot indicates the 50th percentile and here if we turn the screen then we can see the plot which is symmetrical to each other.
